```python
import pandas as pd
```

```python
cars_df = pd.read_csv("cars.csv")
cars_df
```

```python
# First lets take a look at what the columns are
cars_df.columns
```

```python
# lets take a look at the columnms that dont have the feature
cars_df.iloc[:,:15]
```

```python
# this generates 1220 columns from 15 columns... Do not want to do this
pd.get_dummies(cars_df.iloc[:,:15])
```

```python
#lets use a label encoder so we dont create so many columns
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(cars_df["manufacturer_name"])
```

```python
encoded_name = le.transform(cars_df["manufacturer_name"])
encoded_name
```

```python
inverse_encoded_name = le.inverse_transform(encoded_name)
```

```python
# i only want to encode the categorical columns, so i filter those outs.
categorical_columns = cars_df[cars_df.columns[cars_df.dtypes == 'O']]
categorical_columns
```

```python
# encoding the categorical columns
cat_encode_col = categorical_columns.apply(le.fit_transform)
cat_encode_col
```

```python
# getting columns that are numerical
numerical_columns = cars_df[cars_df.columns[cars_df.dtypes != 'O']] # checks if the datatype is not a string
numerical_columns
```

```python
# combining the columns that are numerical and encoded
pd.concat([cat_encode_col,numerical_columns], axis=1)
```
